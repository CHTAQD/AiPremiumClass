```python
import numpy as np
import matplotlib.pyplot as plt

# 目标函数 f(x) = x^2
def f(x):
    return x**2

# 计算梯度（导数） f'(x) = 2x
def gradient(x):
    return 2 * x

# 梯度下降参数
learning_rate = 0.1  # 学习率，决定每次走多远
x = 5.0  # 初始值，假设从 x=5 开始
iterations = 20  # 迭代次数
history = [x]  # 记录下降过程

# 梯度下降循环
for _ in range(iterations):
    x = x - learning_rate * gradient(x)  # 更新 x
    history.append(x)  # 记录 x 值

# 画出下降过程
x_vals = np.linspace(-6, 6, 100)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = x^2")  # 画出函数曲线
plt.scatter(history, [f(i) for i in history], color='red')  # 画出下降过程的点
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("梯度下降示例")
plt.legend()
plt.show()
```
